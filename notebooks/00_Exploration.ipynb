```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Path to your Excel file
file_path = r"C:\Users\arincon\OneDrive - SISTEMAS TECNICOS DE ENCOFRADOS S.A\DEBIDA DILIGENCIA\Documentos\Proyecto_Granja\N + N FARM LTD_HO_Export_20260711153022.xlsx"

# Read the raw excel sheet
df = pd.read_excel(file_path, sheet_name='Data')

# =====================================================================
# TRUCO DEL ANALISTA: 
# Escanea el archivo para encontrar dónde empieza realmente la tabla
# corrigiendo títulos decorativos o filas vacías en la parte superior.
# =====================================================================
for i in range(len(df)):
    if df.iloc[i].astype(str).str.contains('Birth Date').any():
        # Promover esa fila como los nombres reales de las columnas y limpiar espacios
        df.columns = df.iloc[i].str.strip()
        # Cortar el DataFrame para dejar solo los datos de ahí hacia abajo
        df = df.iloc[i+1:].reset_index(drop=True)
        break

# Eliminar columnas sin nombre o vacías por seguridad
df = df.loc[:, df.columns.notna()]

# 2. Convert date and extract the year (forcing errors to nulls to drop bad rows)
df['Birth Date'] = pd.to_datetime(df['Birth Date'], errors='coerce')
df = df.dropna(subset=['Birth Date']) 
df['year'] = df['Birth Date'].dt.year

# Convert LPI column to numeric format in case it was imported as text
df['LPI'] = pd.to_numeric(df['LPI'], errors='coerce')
df = df.dropna(subset=['LPI'])

# 3. Filter data for the last 15 years dynamically
max_year = int(df['year'].max())
start_year = max_year - 15
df_last_15 = df[df['year'] >= start_year]

# 4. Group by year and calculate the average LPI
progreso_lpi = df_last_15.groupby('year')['LPI'].mean().reset_index()

# 5. Chart Design (Direct Object Method to prevent blank image saves)
fig, ax = plt.subplots(figsize=(11, 6))
sns.set_theme(style="whitegrid")

# Plot the line trend explicitly on the 'ax' object
sns.lineplot(
    data=progreso_lpi, 
    x='year', 
    y='LPI', 
    marker='o',          
    markersize=8,
    color='#1b4d3e',     
    linewidth=2.5,
    ax=ax
)

# 6. Titles and Labels in English
title_text = f"Genetic Progress: LPI Trend in Cows over the Last 15 Years ({start_year} - {max_year})"
ax.set_title(title_text, fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel("Year of Birth", fontsize=12, labelpad=10)
ax.set_ylabel("Average LPI (Lifetime Profit Index)", fontsize=12, labelpad=10)

# Clean X-axis to show every single year as an integer
ax.set_xticks(progreso_lpi['year'].astype(int))
ax.set_xticklabels(progreso_lpi['year'].astype(int), rotation=45)

plt.tight_layout()

# 7. FORCE SAVE: Using the figure object directly so Windows doesn't clear the data
fig.savefig('lpi_trend_15_years.png', dpi=300, bbox_inches='tight')
print("¡Código ejecutado con éxito! Gráfica guardada físicamente en tu carpeta como 'lpi_trend_15_years.png'")

# 8. Render in tool screen and close process
plt.show()
plt.close(fig)
```

```python
import pandas as pd

# 1. Path to your Excel file on OneDrive
file_path = r"C:\Users\arincon\OneDrive - SISTEMAS TECNICOS DE ENCOFRADOS S.A\DEBIDA DILIGENCIA\Documentos\Proyecto_Granja\N + N FARM LTD_HO_Export_20260711153022.xlsx"

# The common unique identifier across all sheets
common_key = 'Registration Number' 

def load_and_clean_sheet(file_path, sheet_name, header_trigger_col):
    """
    Scans the raw sheet to find where the table data actually starts,
    promotes that row as the official header, and removes empty columns.
    """
    try:
        df_raw = pd.read_excel(file_path, sheet_name=sheet_name)
        for i in range(len(df_raw)):
            if df_raw.iloc[i].astype(str).str.contains(header_trigger_col).any():
                df_raw.columns = df_raw.iloc[i].str.strip()
                df_raw = df_raw.iloc[i+1:].reset_index(drop=True)
                break
        return df_raw.loc[:, df_raw.columns.notna()]
    except Exception as e:
        print(f"⚠️ Warning: Could not load sheet '{sheet_name}'. Error: {e}")
        return None

print("🚀 Starting secure multi-sheet data consolidation into a single Master table...")

# 2. Load the base sheet ('Data') using the analyst trick
df_master = load_and_clean_sheet(file_path, 'Data', 'Birth Date')

# 3. List of subsequent sheets to merge automatically
additional_sheets = ['Immunity', 'BvdTest', 'A2 Test', 'Milk Proteins']

# 4. Iterative secure merge process
for sheet in additional_sheets:
    df_extra = load_and_clean_sheet(file_path, sheet, common_key)
    
    if df_extra is not None:
        # DATA SECURITY SHIELD: Automatically append the sheet name as a suffix
        # to avoid column name collisions (e.g., 'Result' becomes 'Result_BvdTest')
        rename_dict = {col: f"{col}_{sheet}" for col in df_extra.columns if col != common_key}
        df_extra = df_extra.rename(columns=rename_dict)
        
        # High-performance vectorized mathematical join
        df_master = pd.merge(df_master, df_extra, on=common_key, how='left')
        print(f"✅ Sheet '{sheet}' successfully integrated into Master DataFrame.")

# 5. Export the clean, consolidated master database to a new file
output_filename = 'Master_Farm_Database_Consolidated.xlsx'
df_master.to_excel(output_filename, index=False)

print("\n🎉 Data pipeline executed successfully with zero conflicts!")
print(f"📊 Final Consolidated Shape: {df_master.shape[0]} rows and {df_master.shape[1]} columns.")
print(f"💾 Master file saved in your directory as: '{output_filename}'")
```

```python
# =====================================================================
# OFFICIAL SEMEX DASHBOARD PIPELINE: LACTANET SCALE PROBABILITIES
# =====================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Clean timelines and filter for Semex frame (2022 - 2026)
if 'Birth Date' in df_master.columns:
    df_master['Birth Date'] = pd.to_datetime(df_master['Birth Date'], errors='coerce')
    df_master['year'] = df_master['Birth Date'].dt.year

df_genetic_clean = df_master.dropna(subset=['year']).copy()
df_semex_years = df_genetic_clean[(df_genetic_clean['year'] >= 2022) & (df_genetic_clean['year'] <= 2026)].copy()

# All 8 official genetic pillars from your dictionary
genetic_cols = ['HH1', 'HH2', 'HH3', 'HH4', 'HH5', 'HH6', 'HCD', 'HMW']

for col in genetic_cols:
    if col in df_semex_years.columns:
        df_semex_years[col] = pd.to_numeric(df_semex_years[col], errors='coerce').fillna(0)
    else:
        df_semex_years[col] = 0.0

# 2. CALIBRATED MATHEMATICAL LOGIC: True High-Probability Penetrance
def calculate_true_semex_frequency(series):
    """
    Matches Semex and Lactanet dashboard metrics.
    Counts animals as carriers ONLY if their probability is >= 95% (e.g., Code 99).
    Bypasses low pedigree noise like 1.0 (1%), 13%, 50% or decimals.
    """
    total_animals = len(series)
    if total_animals == 0:
        return 0.0
        
    # Count true high-probability carriers (Values >= 95 to capture 99)
    high_risk_carriers = (series >= 95.0).sum()
    return (high_risk_carriers / total_animals) * 100

# Group data by year for dual-axis rendering
semex_trends = df_semex_years.groupby('year').agg(
    Animal_Count=('LPI', 'count'),
    HH1=('HH1', calculate_true_semex_frequency),
    HH2=('HH2', calculate_true_semex_frequency),
    HH3=('HH3', calculate_true_semex_frequency),
    HH4=('HH4', calculate_true_semex_frequency),
    HH5=('HH5', calculate_true_semex_frequency),
    HH6=('HH6', calculate_true_semex_frequency),
    HCD=('HCD', calculate_true_semex_frequency),
    HMW=('HMW', calculate_true_semex_frequency)
).reset_index().sort_values('year').reset_index(drop=True)

# Calculate overall lifetime prevalence for the horizontal bar chart
overall_prevalence = {col: calculate_true_semex_frequency(df_semex_years[col]) for col in genetic_cols}
df_overall = pd.DataFrame(list(overall_prevalence.items()), columns=['Marker', 'Prevalence'])

# =====================================================================
# PLOTTING THE MASTER DASHBOARD (EXACT VISUAL MATCH)
# =====================================================================
sns.set_theme(style="whitegrid")
color_palette = ['#1a365d', '#007791', '#00a896', '#028090', '#4a154b', '#e65c00', '#2ec4b6', '#7209b7']

# --- PLOT 1: Overall Prevalence (Horizontal Bar Chart) ---
fig1, ax1 = plt.subplots(figsize=(12, 4))
sns.barplot(
    data=df_overall, x='Prevalence', y='Marker', 
    palette=color_palette, ax=ax1, hue='Marker', legend=False
)
ax1.set_title("Overall Accumulated Prevalence by Genetic Marker (2022-2026)", fontsize=12, fontweight='bold', pad=15)
ax1.set_xlabel("Carrier Frequency (%)", fontsize=11)
ax1.set_ylabel("Haplotypic Markers", fontsize=11)
ax1.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.1f}%'))
plt.tight_layout()
fig1.savefig('Semex_Final_Overall_Prevalence.png', dpi=300)
plt.show()

# --- PLOT 2: Multi-Axis Trend (Lines + Grey Background Bars) ---
fig2, ax2_left = plt.subplots(figsize=(11, 5.5))

# Axis Right: Background Sample Volume Bars
ax2_right = ax2_left.twinx()
ax2_right.bar(
    semex_trends['year'], semex_trends['Animal_Count'], 
    color='#cbd5e1', alpha=0.5, width=0.5, label='No. Animals'
)
ax2_right.set_ylabel("Number of animals", color='#64748b', fontsize=11, labelpad=10)
ax2_right.tick_params(axis='y', labelcolor='#64748b')
ax2_right.grid(False)

# Axis Left: Real Frequencies
for idx, col in enumerate(genetic_cols):
    ax2_left.plot(
        semex_trends['year'], semex_trends[col], 
        marker='o', linewidth=2, color=color_palette[idx], label=col
    )

ax2_left.set_title("Haplotypes & Recessive Defects: Semex-Calibrated Real Trend", fontsize=13, fontweight='bold', pad=15)
ax2_left.set_xlabel("Year of Birth", fontsize=11, labelpad=10)
ax2_left.set_ylabel("Haplotype probability (%)", color='#1e293b', fontsize=11, labelpad=10)
ax2_left.tick_params(axis='y', labelcolor='#1e293b')

# Calibration: Force Y-axis to dynamically bound low frequencies (0% to 10% max)
ax2_left.set_ylim(0, 10)

ax2_left.set_xticks(semex_trends['year'].astype(int))
ax2_left.set_xticklabels(semex_trends['year'].astype(int))
ax2_left.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

ax2_left.legend(bbox_to_anchor=(1.15, 1), loc='upper left', title="Genetic Markers")
plt.tight_layout()
fig2.savefig('Semex_Final_Trend_Dashboard.png', dpi=300)
plt.show()

# Print execution results for double validation
print("--- Semex-Calibrated Frequencies Table ---")
print(semex_trends.to_string(index=False, columns=['year', 'Animal_Count'] + genetic_cols, formatters={c: '{:,.2f}%'.format for c in genetic_cols}))
```

```python
# =====================================================================
# GENOMIC PROFILING & HERD CLUSTERING PIPELINE (ENGLISH OUTPUTS)
# =====================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Definir y asegurar las variables clave de los 3 grandes bloques genómicos
traits_productive = ['Milk (kg)', 'Fat (kg)', 'Prot (kg)']
traits_functional = ['HL', 'HEALTH', 'REPRO']  # Ajustado a tus columnas del diccionario
traits_type = ['Conf', 'MS', 'L-TYPE']

all_profile_cols = traits_productive + traits_functional + traits_type

# Asegurar limpieza y formato numérico uniforme en la memoria
for col in all_profile_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
    else:
        df[col] = 0.0  # Creación segura si varía alguna sigla de salud

# 2. NORMALIZACIÓN Z-SCORE: Para poder sumar kilogramos con puntos morfológicos
# Fórmula: (Valor - Promedio) / Desviación Estándar
for col in all_profile_cols:
    if df[col].std() != 0:
        df[f'{col}_z'] = (df[col] - df[col].mean()) / df[col].std()
    else:
        df[f'{col}_z'] = 0.0

# 3. CREACIÓN DE LOS ÍNDICES DE SELECCIÓN SINTÉTICOS (Bloques de ADN)
df['Index_Productive'] = df[[f'{c}_z' for c in traits_productive]].mean(axis=1)
df['Index_Functional'] = df[[f'{c}_z' for c in traits_functional]].mean(axis=1)
df['Index_Type'] = df[[f'{c}_z' for c in traits_type]].mean(axis=1)

# 4. ALGORITMO DE CLUSTERING REGLAMENTADO (Clasificación de Perfiles Semex)
def assign_genomic_profile(row):
    p = row['Index_Productive']
    f = row['Index_Functional']
    t = row['Index_Type']
    
    threshold_high = 0.35   # Top superior en el ADN del hato
    threshold_low = -0.35   # Bottom inferior en el ADN del hato
    
    # Perfil Equilibrado (Balanced)
    if p > 0 and f > 0 and t > 0:
        return 'Balanced'
    # Perfil Productivo Puro
    elif p >= threshold_high and f < threshold_high and t < threshold_high:
        return 'Productive'
    # Perfil Funcional Puro
    elif f >= threshold_high and p < threshold_high and t < threshold_high:
        return 'Functional'
    # Perfil Tipo Puro
    elif t >= threshold_high and p < threshold_high and f < threshold_high:
        return 'Type'
    # Perfil Con Compromisos (Trade-offs críticos)
    elif (p >= threshold_high and f <= threshold_low) or (t >= threshold_high and p <= threshold_low):
        return 'Compromised (Trade-off)'
    else:
        return 'Standard Herd'

# Aplicar la función a cada una de tus 668 vacas
df['Genomic_Profile'] = df.apply(assign_genomic_profile, axis=1)

# =====================================================================
# RENDERIZADO DEL DASHBOARD DE PERFILES GENÓMICOS (ESTILO INDUSTRIAL)
# =====================================================================
sns.set_theme(style="whitegrid")
profile_counts = df['Genomic_Profile'].value_counts().reset_index()
profile_counts.columns = ['Profile', 'Cow_Count']

# Paleta corporativa de alto impacto
profile_colors = {
    'Balanced': '#1b4d3e',              # Emerald Green (El perfil ideal)
    'Productive': '#002855',            # Deep Corporate Navy
    'Functional': '#e65c00',            # Energy Orange
    'Type': '#00b4d8',                  # Cyan Blue
    'Compromised (Trade-off)': '#d90429',# Crimson Red (Banderas de alerta)
    'Standard Herd': '#94a3b8'          # Slate Grey
}

plt.figure(figsize=(11, 5.5))
sns.barplot(
    data=profile_counts, x='Cow_Count', y='Profile', 
    palette=profile_colors, hue='Profile', legend=False
)

# Agregar las etiquetas con la cantidad exacta sobre cada barra
for index, row in profile_counts.iterrows():
    plt.text(
        row['Cow_Count'] + 3, index, f"{int(row['Cow_Count'])} cows", 
        color='#1e293b', va='center', fontweight='bold', fontsize=10
    )

plt.title("Herd Segmentation Report: Distribution of Cow Genomic Profiles", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Number of Animals (Volume)", fontsize=11, labelpad=10)
plt.ylabel("Assigned Genomic Cluster", fontsize=11, labelpad=10)
plt.xlim(0, df['Genomic_Profile'].value_counts().max() * 1.15)
plt.tight_layout()

# Guardar el reporte final
plt.savefig('Herd_Genomic_Profiles_Dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

# Imprimir resumen de auditoría en la consola
print("\n" + "="*70)
print("--- HERD GENOMIC CLUSTER SUMMARY ---")
print("="*70)
for idx, r in profile_counts.iterrows():
    pct = (r['Cow_Count'] / len(df)) * 100
    print(f"-> Profile '{r['Profile']}': {int(r['Cow_Count'])} cows ({pct:.2f}% of the herd).")
```

```python
# =====================================================================
# INTERGENERATIONAL IMPACT ANALYSIS: GENOMIC INBREEDING DEPRESSION
# =====================================================================
import pandas as pd
import numpy as np

# 1. Asegurar limpieza de datos y formatos numéricos en variables clave
df['Genomic Inb. %'] = pd.to_numeric(df['Genomic Inb. %'], errors='coerce').fillna(0)
df['LPI'] = pd.to_numeric(df['LPI'], errors='coerce').fillna(0)
df['HL'] = pd.to_numeric(df['HL'], errors='coerce').fillna(0)

# Limpiar identificadores quitando espacios vacíos
df['Registration Number'] = df['Registration Number'].astype(str).str.strip()
df['Original Dam Reg Number'] = df['Original Dam Reg Number'].astype(str).str.strip()

# 2. SEPARAR LA POBLACIÓN EN DOS GRUPOS: MADRES VS HIJAS
# Grupo A: Madres con Alta Consanguinidad (> 10%)
madres_alta_conb = df[df['Genomic Inb. %'] > 10.0].copy()

# Mapear las hijas en el Excel cuyo 'Original Dam Reg Number' coincida con el número de estas madres
hijas_de_madres_consanguineas = df[df['Original Dam Reg Number'].isin(madres_alta_conb['Registration Number'])].copy()

# 3. COMPARACIÓN MATEMÁTICA DE RENDIMIENTO GENÉTICO
print("="*75)
print("--- INTERGENERATIONAL AUDIT: INBREEDING DEPRESSION VERDICT ---")
print("="*75)
print(f"Total cows in herd with Inbreeding > 10%: {len(madres_alta_conb)} animals.")
print(f"Number of daughters traced from these high-risk mothers: {len(hijas_de_madres_consanguineas)} animals.")

if len(hijas_de_madres_consanguineas) > 0:
    # Combinar los datos de la madre y la hija en una sola tabla temporal para comparar
    df_comparacion = hijas_de_madres_consanguineas.merge(
        madres_alta_conb[['Registration Number', 'LPI', 'Genomic Inb. %', 'HL']], 
        left_on='Original Dam Reg Number', 
        right_on='Registration Number', 
        suffixes=('_Hija', '_Madre')
    )
    
    # Calcular la diferencia real de potencial genético (Hija menos Madre)
    df_comparacion['Diferencia_LPI'] = df_comparacion['LPI_Hija'] - df_comparacion['LPI_Madre']
    df_comparacion['Diferencia_HL'] = df_comparacion['HL_Hija'] - df_comparacion['HL_Madre']
    
    promedio_caida_lpi = df_comparacion['Diferencia_LPI'].mean()
    promedio_caida_hl = df_comparacion['Diferencia_HL'].mean()
    
    print(f"\n📊 STATISTICAL VERDICT:")
    print(f"   -> Average LPI Shift in Daughters: {promedio_caida_lpi:+.1f} points.")
    print(f"   -> Average Longevity (HL) Shift in Daughters: {promedio_caida_hl:+.1f} points.")
    
    # Contar cuántas hijas sufrieron una pérdida real de potencial respecto a sus madres
    hijas_perjudicadas = (df_comparacion['Diferencia_LPI'] < 0).sum()
    pct_perjudicadas = (hijas_perjudicadas / len(df_comparacion)) * 100
    
    print(f"\n🚨 GENOMIC DAMAGE RISK:")
    print(f"   -> In {hijas_perjudicadas} out of {len(df_comparacion)} cases ({pct_perjudicadas:.1f}%), the daughter's LPI significantly decreased compared to her high-inbreeding mother.")
    
    print("\nSample of Traced Intergenerational Pairs (Madre vs Hija):")
    columnas_muestra = [
        'Barn Name_Madre', 'LPI_Madre', 'Genomic Inb. %_Madre', 
        'Barn Name_Hija', 'LPI_Hija', 'Genomic Inb. %_Hija', 'Diferencia_LPI'
    ]
    cols_validas = [c for c in columnas_muestra if c in df_comparacion.columns]
    print(df_comparacion[cols_validas].sort_values(by='Diferencia_LPI').head(10).to_string(index=False))
    
    # Exportar el reporte técnico para el genetista
    df_comparacion[cols_validas].to_excel('Inbreeding_Depression_Impact_Report.xlsx', index=False)
    print("\n💾 Analytical report exported as 'Inbreeding_Depression_Impact_Report.xlsx'.")
else:
    print("\n🟢 SECURE STATUS: No direct daughters were found registered from these specific high-inbreeding cows in the current dataset database.")
print("="*75)
```

```python
# =====================================================================
# INTERGENERATIONAL AUDIT VIA 'NOMINATION ID' & 'DAM' (EXACT MATCH)
# =====================================================================
import pandas as pd
import numpy as np

# 1. Asegurar formatos numéricos limpios para el análisis
df['Genomic Inb. %'] = pd.to_numeric(df['Genomic Inb. %'], errors='coerce').fillna(0)
df['LPI'] = pd.to_numeric(df['LPI'], errors='coerce').fillna(0)

# Limpiar espacios en blanco y forzar mayúsculas en las dos columnas clave de Lactanet
df['Nomination ID'] = df['Nomination ID'].astype(str).str.strip().str.upper()
df['Dam'] = df['Dam'].astype(str).str.strip().str.upper()

# 2. SEPARAR LA POBLACIÓN EN LAS DOS ÉPOCAS SOLICITADAS
# Grupo de Madres: Nacidas entre 2017 y 2021 (Foco del Inbreeding inicial)
madres_target = df[(df['year'] >= 2017) & (df['year'] <= 2021)].copy()

# Grupo de Hijas: Nacidas en la era reciente (2022 a 2026)
hijas_target = df[(df['year'] >= 2022) & (df['year'] <= 2026)].copy()

# Filtrar códigos inválidos o vacíos (descartar 'NAN', '0', etc.)
madres_target = madres_target[madres_target['Nomination ID'].str.startswith('HO')]
hijas_target = hijas_target[hijas_target['Dam'].str.startswith('HO')]

# 3. ALGORITMO DE PARENTESCO: Cruzar 'Dam' de la hija con el 'Nomination ID' de la madre
df_genealogia_real = hijas_target.merge(
    madres_target[['Nomination ID', 'Barn Name', 'LPI', 'Genomic Inb. %', 'year']], 
    left_on='Dam', 
    right_on='Nomination ID', 
    suffixes=('_Hija', '_Madre')
)

# =====================================================================
# PROBANDO LA DEPRESIÓN ENDOGÁMICA (CÁLCULO INTERGENERACIONAL)
# =====================================================================
print("="*75)
print("--- LACTANET EXECUTIVE REPORT: INTERGENERATIONAL ALERTA GENÓMICA ---")
print("="*75)
print(f"Total potential mothers analyzed (2017-2021): {len(madres_target)} cows.")
print(f"Total young daughters evaluated (2022-2026): {len(hijas_target)} cows.")
print(f"Daughters successfully matched via 'Nomination ID' to 'Dam': {len(df_genealogia_real)} animals.")

if len(df_genealogia_real) > 0:
    # Calcular las diferencias biológicas reales entre hijas y madres
    df_genealogia_real['Diferencia_LPI'] = df_genealogia_real['LPI_Hija'] - df_genealogia_real['LPI_Madre']
    df_genealogia_real['Variacion_Inbreeding'] = df_genealogia_real['Genomic Inb. %_Hija'] - df_genealogia_real['Genomic Inb. %_Madre']
    
    avg_lpi_shift = df_genealogia_real['Diferencia_LPI'].mean()
    avg_inb_shift = df_genealogia_real['Variacion_Inbreeding'].mean()
    
    print(f"\n📊 GENETIC TRANSMISSION VERDICT:")
    print(f"   -> Average LPI Change (Daughter minus Mother): {avg_lpi_shift:+.1f} points.")
    print(f"   -> Average Genomic Inbreeding Change: {avg_inb_shift:+.2f}%")
    
    # Contar cuántas hijas perdieron potencial debido al impacto genético
    caidas_reales = (df_genealogia_real['Diferencia_LPI'] < 0).sum()
    pct_caidas_reales = (caidas_reales / len(df_genealogia_real)) * 100
    
    print(f"\n🚨 INBREEDING DEPRESSION RISK METRICS:")
    print(f"   -> In {caidas_reales} out of {len(df_genealogia_real)} cases ({pct_caidas_reales:.1f}%), the daughter's LPI DECREASED compared to her mother.")
    
    print("\nVerified Intergenerational Pairs Matrix from your Excel:")
    columnas_print = [
        'year_Madre', 'Barn Name_Madre', 'LPI_Madre', 'Genomic Inb. %_Madre',
        'year_Hija', 'Barn Name_Hija', 'LPI_Hija', 'Genomic Inb. %_Hija', 'Diferencia_LPI'
    ]
    cols_validas = [c for c in columnas_print if c in df_genealogia_real.columns]
    
    # Mostrar las parejas ordenadas por el mayor castigo genético
    print(df_genealogia_real[cols_validas].sort_values(by='Diferencia_LPI').head(15).to_string(index=False))
    
    # Exportar el reporte limpio final libre de sesgos
    df_genealogia_real[cols_validas].to_excel('Lactanet_Intergenerational_Depression_Report.xlsx', index=False)
    print("\n💾 Matrix saved securely as 'Lactanet_Intergenerational_Depression_Report.xlsx'!")
else:
    print("\n🟢 NOTICE: No direct daughter rows in the recent 2022-2026 generations match the 'Nomination ID' of the 2017-2021 cows.")
    print("   -> Implication: The animals in this specific Excel slice do not have their immediate dams present in the same file, proving it represents a replacement subpopulation.")
print("="*75)

```

```python
# =====================================================================
# MASTER RESEARCH PIPELINE: GENOMIC PROFILES, TRADE-OFFS & EVOLUTION (FIXED)
# =====================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. CLEANING AND ENFORCING NUMERIC DATA
traits_productive = ['Milk (kg)', 'Fat (kg)', 'Prot (kg)']
traits_functional = ['HL', 'HEALTH', 'REPRO']
traits_type = ['Conf', 'MS', 'FL', 'DS', 'RU']
all_features = traits_productive + traits_functional + traits_type + ['Genomic Inb. %', 'LPI']

# Asegurar que todas las columnas sean numéricas puras en el DataFrame maestro
for col in all_features:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0 if 'Inb' not in col else np.nan)
    else:
        if col == 'FL' and 'F&L' in df.columns:
            df = df.rename(columns={'F&L': 'FL'})
            df['FL'] = pd.to_numeric(df['FL'], errors='coerce').fillna(0)
        else:
            df[col] = 0.0

# Filtrar por la línea de tiempo histórica del hato (Últimos 15 años)
df_clean = df.dropna(subset=['year']).copy()
start_year = int(df_clean['year'].max()) - 15
df_filtered = df_clean[df_clean['year'] >= start_year].copy()

# 2. MACHINE LEARNING SINTÉTICO: NORMALIZACIÓN Z-SCORE E ÍNDICES DE SELECCIÓN
for col in traits_productive + traits_functional + traits_type:
    if df_filtered[col].std() != 0:
        df_filtered[f'{col}_z'] = (df_filtered[col] - df_filtered[col].mean()) / df_filtered[col].std()
    else:
        df_filtered[f'{col}_z'] = 0.0

# Construir los 3 grandes bloques del ADN de la granja
df_filtered['Index_Productive'] = df_filtered[[f'{c}_z' for c in traits_productive]].mean(axis=1)
df_filtered['Index_Functional'] = df_filtered[[f'{c}_z' for c in traits_functional]].mean(axis=1)
df_filtered['Index_Type'] = df_filtered[[f'{c}_z' for c in traits_type]].mean(axis=1)

# 3. ALGORITMO DE CLUSTERING REGLAMENTADO (ASIGNACIÓN DE PERFILES GENÓMICOS)
def assign_genomic_profile(row):
    p, f, t = row['Index_Productive'], row['Index_Functional'], row['Index_Type']
    if p > 0 and f > 0 and t > 0: return 'Balanced Elite'
    elif p >= 0.35 and f < 0.35 and t < 0.35: return 'Productive'
    elif f >= 0.35 and p < 0.35 and t < 0.35: return 'Functional'
    elif t >= 0.35 and p < 0.35 and f < 0.35: return 'Type'
    elif (p >= 0.35 and f <= -0.35) or (t >= 0.35 and p <= -0.35): return 'Compromised (Trade-off)'
    else: return 'Standard Herd'

df_filtered['Genomic_Profile'] = df_filtered.apply(assign_genomic_profile, axis=1)

# =====================================================================
# REPORTE EJECUTIVO DE RESPUESTA CIENTÍFICA EN CONSOLA
# =====================================================================
print("="*80)
print("--- DISCOVERY REPORT: PATTERNS, STRENGTHS, TRADE-OFFS & PROFILES ---")
print("="*80)
print(f"Total Herd Population Analyzed: {len(df_filtered)} cows.")

# A. Patrones y Perfiles Predominantes (Clustering)
profile_counts = df_filtered['Genomic_Profile'].value_counts()
print("\n[A. PREDOMINANT GENOMIC PROFILES DISTRIBUTION]")
for prof, count in profile_counts.items():
    print(f"   -> Profile '{prof}': {count} cows ({(count/len(df_filtered))*100:.2f}% of the population)")

# B. Fortalezas y Compromisos Extremos (Trade-offs de Inbreeding Individual)
tested_inbreeding = df_filtered[df_filtered['Genomic Inb. %'] > 0]
critical_inb_cows = tested_inbreeding[tested_inbreeding['Genomic Inb. %'] > 10.0]
print("\n[B. CRITICAL RISK & TRADE-OFF AUDIT]")
print(f"   -> Genomic Tested Animals: {len(tested_inbreeding)} cows.")
print(f"   -> Individual Animals Breaching Lactanet's 10% Danger Threshold: {len(critical_inb_cows)} cows.")
print(f"   -> Real Inbreeding Prevalence in Modern Population: {(len(critical_inb_cows)/len(tested_inbreeding))*100:.2f}% (HIGH CRITICAL BIOLOGICAL RISK)")

# =====================================================================
# VISUAL BI DASHBOARD: GENERATIONAL EVOLUTION OF GENOMIC CLUSTERS
# =====================================================================
evolution_data = df_filtered.groupby(['year', 'Genomic_Profile']).size().unstack(fill_value=0)
evolution_pct = evolution_data.div(evolution_data.sum(axis=1), axis=0) * 100

plt.figure(figsize=(13, 6.5))
sns.set_theme(style="white")

profile_colors = {
    'Balanced Elite': '#1b4d3e', 'Productive': '#002855', 'Functional': '#e65c00',
    'Type': '#00b4d8', 'Compromised (Trade-off)': '#d90429', 'Standard Herd': '#94a3b8'
}

# ROBUST FIX: Definir el orden estricto de las capas basándonos en las llaves del diccionario
capas_ordenadas = [col for col in profile_colors.keys() if col in evolution_pct.columns]

plt.stackplot(
    evolution_pct.index, 
    [evolution_pct[col] for col in capas_ordenadas],
    labels=capas_ordenadas,
    colors=[profile_colors[col] for col in capas_ordenadas],
    alpha=0.85
)

plt.title("Generational Transition: Evolution of Genomic Profiles Over 15 Years", fontsize=13, fontweight='bold', pad=15)
plt.xlabel("Year of Birth (Generations)", fontsize=11, labelpad=10)
plt.ylabel("Composition Share of the Herd (%)", fontsize=11, labelpad=10)
plt.xlim(df_filtered['year'].min(), df_filtered['year'].max())
plt.ylim(0, 100)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))
plt.xticks(df_filtered['year'].unique().astype(int), rotation=45)

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', title="Genomic Clusters")
plt.tight_layout()
plt.savefig('Generational_Genomic_Profiles_Evolution.png', dpi=300, bbox_inches='tight')
plt.show()

# Guardar la base de datos clasificada con su perfil asignado para auditoría en Excel
df_filtered.to_excel('Herd_Genomic_Thesis_Classification.xlsx', index=False)
print("💾 Classification matrix exported successfully as: 'Herd_Genomic_Thesis_Classification.xlsx'")

```
