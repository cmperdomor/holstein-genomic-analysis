```python
# ============================================================
# IMPORT REQUIRED LIBRARIES
# ============================================================

import pandas as pd
import numpy as np
```

```python
# ============================================================
# LOAD THE DATASET
# ============================================================

file_path = "../Master_Database_Consolidated_Final.xlsx"

df = pd.read_excel(file_path)

print("Dataset loaded successfully!")
```

```python
# ============================================================
# BASIC INFORMATION
#how many lines and columns my file has
# ============================================================

print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
# ============================================================
# DISPLAY THE FIRST ROWS
#Verify that the correct dataset was loaded
# ============================================================

df.head()
```

```python
# ============================================================
# DATASET AUDIT
# Understanding the structure before cleaning
# ============================================================

print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)

print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

print("\nData types:")
display(df.dtypes)

print("\nMissing values:")
display(df.isna().sum().sort_values(ascending=False).head(20))
df.isna().sum()
```

```python
# ============================================================
# COLUMN NAMES
# ============================================================

print(df.columns.tolist())
```

```python
# ============================================================
# DUPLICATED ROWS
# ============================================================

print(f"Duplicated rows: {df.duplicated().sum()}")
```

```python
# ============================================================
# DATASET DIMENSIONS
# How large is our dataset?
# ============================================================

print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
# ============================================================
# DATA TYPES
# Did pandas correctly identify each variable?
# ============================================================

display(df.dtypes)
```

```python
# ============================================================
# MISSING VALUES
# Which variables contain missing information?
# ============================================================

missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)

display(missing)
```

```python
# ============================================================
# DUPLICATED ROWS
# ============================================================

duplicates = df.duplicated().sum()

print(f"Duplicated rows: {duplicates}")
```

```python
# ============================================================
# COLUMN NAMES
# ============================================================

for column in df.columns:
    print(column)
```

```python
missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)
```

```python
# ============================================================
# DATA QUALITY SUMMARY
# ============================================================

quality = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Missing (%)": round(df.isnull().mean()*100,2),
    "Data Type": df.dtypes
})

quality = quality.sort_values(
    by="Missing (%)",
    ascending=False
)

quality.head(25)
```

```python
# ============================================================
# GENOMIC VARIABLES
# ============================================================

genomic_columns = [
    col
    for col in df.columns
    if any(word in col.upper() for word in
           ["MILK","FAT","PROT","LPI","PRO","HL","MS","FL","DS","RU","CONF"])
]

genomic_columns
```

```python
# ============================================================
# DATA QUALITY SUMMARY
# ============================================================

quality = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Missing (%)": round(df.isnull().mean()*100,2),
    "Data Type": df.dtypes
})

quality = quality.sort_values(
    by="Missing (%)",
    ascending=False
)

quality.head(25)
```
